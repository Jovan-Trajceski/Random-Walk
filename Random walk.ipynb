# Random Walk Problem

You are walking up the Emire State building to XYZ Inc. HeadQuarters and playing a game with a friend.

You throw a dice one hundred times:
* if it is 1 or 2 you will go one step down
* if it is 3, 4, or 5 you will go one step up
* if you throw a 6, you will throw the dice again and will walk up the resulting number of steps
* you cant go lower than step number 0
* you admit you are a bit clumsy and have a chance of 0.1% of falling down the stairs when you make a move
* falling down means that you have to start again from step 0

# Question

With all this in mind you bet with your friend that you will reach 60 steps high

* What is the chance you will win this bet?

# Solution

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate random walk 10,000 times
all_walks = []
for i in range(10000) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.figure(figsize=(15,10))
plt.plot(np_aw_t, marker = 'o', linestyle = ':')
plt.grid(axis='y', alpha=0.75,linestyle='dotted')
plt.grid(axis='x', alpha=0.75, linestyle='dotted')
plt.xlabel('Walk', fontsize=14)
plt.ylabel('Step', fontsize=14)
plt.title('Random Walk Process - 10,000 Simulations', loc='left',fontsize=20, style='italic', fontweight=1000, 
          color='black')
plt.show()

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.figure(figsize=[15,10])
plt.hist(ends, bins=10,  color = 'green', edgecolor = 'black', rwidth=0.8)
plt.grid(axis='y', alpha=0.75, linestyle='dotted')
plt.xlabel('End Step', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Random Walk Result - 10,000 Simulations', loc='left',fontsize=20, style='italic', fontweight=1000, 
          color='black')
plt.show()

# Print probability of winning the bet
prob=np.mean(ends>=60)*100
print('Probability of winning this bet is: ' + str(round(prob,2)) + '%')
```

# Conclusion

There is a high probability of you winning this bet.
